```python
# Let's count the i's in "mississippi"
count = 0
word = "mississippi"
letter = "i"
for ch in word:
    if ch == letter:
        count = count + 1

print(f"There are {count} {letter}'s in {word}")
```

```python
password = "pickle"
max_attempts = 3
attempts = 0
access_granted = False
locked_out = False
while True:
    pw = input("Enter your password:")
    attempts = attempts + 1
    if pw == password:
        print("Access Granted!")
        access_granted = True
        break
    elif attempts == max_attempts:
        print("You are locked out... too many tries")
        locked_out = True
        break
    else:
        print(f"Invalid password! Attempt {attempts} of {max_attempts}")

if access_granted:
    print("Here's $20")
else:
    print("No Money for you!")
```
