# Part One

You have been tasked with writing a program to help employees of a travel company convert currency.

The company travels to the following places. Next to the place is the exchange rate.

- Europe:  1 US Dollar == 0.9 Euro
- China: 1 US Dollar == 7.2 Chineese Yuan
- Russia: 1 US Dollar == 64.2 Russian Ruble

You should write a program to input the amount of currency in US Dollars, and then input the place (either Europe, China, or Russia). 

The program should then output the appropriate amount of exchanged currency. 

Example Run 1:

```
Enter the amount of Currency in US Dollars: $100
Enter the place you are travelling: Europe, China or Russia? China

$100.00 US Dollars is $720.00 Chineese Yuan

```


## Problem Analysis

Inputs:

    PROMPT 1: What are the inputs?

    - amount_usd float
    - target_curreny str

Outputs:

    PROMPT 2: What are the outputs?

    - exchange_rate
    - amount_usd
    - converted_amount

    
Algorithm (Steps in Program):

    PROMPT 3: What are the steps, from input to output? Express in your own words

    1) with inputs amount_usd float, target_curreny str
    2) if target_currency equals 'canada'
    3)    exchange_rate = 0.7
    4) elif target_currency equals 'russia'
    5)    exchange_rate = 64.2
    6) and so on...
    7) converted_amount = amount_usd * exchange_rate
    8) print amount_usd, exchange_rate, converted_amount
    
    

````python
# PROMPT 4 Code the solution!!!
Euro_rate = 0.9
Yuan_rate = 7.2
Russia_rate = 64.2
Modor_rate = .004

#get user input
usd = float(input("Enter the amount of currency in USD Dollars: $"))
place = input("Enter the place you're traveling: China, Europe or Russia? ")

#convert currency
if place == "Europe":
    converted = usd * Euro_rate
    currency = "Euro"
elif place == "China":
    converted = usd * Yuan_rate
    currency = "Chineese Yuan"
elif place == "Russia":
    converted = usd * Russia_rate
    currency = "Russian Ruble"
elif place == "Modor":
    converted = usd * Modor_rate
    currency = "Gold Coins"

#output this
print(f"${usd:.2f} US dollar is ${converted:.2f} {currency}")
````
