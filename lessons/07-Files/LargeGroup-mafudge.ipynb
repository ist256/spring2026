```python
def tally_votes(poll):
    a,b,c,d = 0,0,0,0
    for response in poll.split():
        if response == "A":
            a = a + 1
        elif response == "B":
            b = b + 1
        elif response == "C":
            c = c + 1
        elif response == "D":
            d = d + 1
        else:
            pass
    return a,b,c,d 
```

```python
poll = read_file_contents("poll-responses.txt")
a,b,c,d = tally_votes(poll)
print(f"""
A = {a}
B = {b}
C = {c}
D = {d}
""")
```

```python
polls = read_file_contents("poll-responses2.txt")
for poll in polls.split("\n"):
    a,b,c,d = tally_votes(poll)
    print(f"""
    A = {a}
    B = {b}
    C = {c}
    D = {d}
    """)
```
