```python
number_list = []
for tnum in text_numbers:
    real_num = int(tnum)
    number_list.append(real_num)

number_list
```

```python
import random
possible_numbers = list(range(1,61))
winning_numbers = []
draw = 7
for i in range(draw):
    # pick number from possible_numbers
    pick = random.choice(possible_numbers)
    # remove it from possible numbers
    possible_numbers.remove(pick)
    # append it to winning_numbers
    winning_numbers.append(pick)

winning_numbers.sort()
print(f"Winning Numbers: {winning_numbers}")
print()
print()
print(possible_numbers)
```
