```python
items = []
while True:
    item = input("Enter grocery item:").lower()
    if item == 'quit':
        break
    if item not in items:
        items.append(item)
        items.sort()
        print(items)
    else:
        print(f"{item} is already on the list")
        
```

```python
def decode_word(stylized_word):
    replacements = {
        '4': 'a',
        '0': 'o',
        '3': 'e',
        '1': 'i',
        '5': 's',
        '7': 't'
    }

    decoded_word = ''.join(replacements.get(char, char) for char in stylized_word)
    return decoded_word.lower()

pw_file = "common-passwords.txt"
with open(pw_file, "r") as f:
    data = f.read()
    passwords = data.split()

print("Password Checker")
while True:
    pw = input("Enter candiate password or 'quit' to stop:")
    if pw == 'quit':
        break
    pw_decoded = decode_word(pw)
    if pw_decoded in passwords:
        print(f"I would not use {pw}, decoded as {pw_decoded}. Too common")
    else:
        print(f"Your password was not on the list")
```
