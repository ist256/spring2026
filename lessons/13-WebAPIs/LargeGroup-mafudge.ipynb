```python
def azure_sentiment(text):
    '''
    this uses the azure api to get the sentiment (pos/neu/neg) for text
    '''
    data = { "text" : text }
    url = 'https://cent.ischool-iot.net/api/azure/sentiment'
    headers = {
        'accept': 'application/json',
        'X-API-KEY': '0f92bf60ff8c50f7e6929e57',
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    response = requests.post(url, headers=headers, params=params, data=data)
    response.raise_for_status()
    data = response.json()
    sentiment = data['results']['documents'][0]
    return sentiment

azure_sentiment("This is possibly the worst pair of sneakers sketchers has ever created. The so un comfortable.")
```

```python
def weather(lat, lon):
    url = 'https://cent.ischool-iot.net/api/weather/current'
    headers = {
        'accept': 'application/json',
        'X-API-KEY': '0f92bf60ff8c50f7e6929e57'
    }
    params = {
        'units': 'imperial',
        'lon': lon,
        'lat': lat
    }

    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    weather = response.json()
    return weather

weather(43,-78)
```

```python
def geocode(location):
    '''
    For google geocode in goes the location, out goes the lat /lon
    '''
    url = 'https://cent.ischool-iot.net/api/google/geocode'
    headers = {
        'accept': 'application/json',
        'X-API-KEY': '0f92bf60ff8c50f7e6929e57'
    }
    params = {
        'location': location
    }
    
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    location = response.json()
    return location

location = geocode("JMA Wireless Dome")
location
```

```python
import requests
from IPython.display import display
from ipywidgets import interact_manual

@interact_manual(location="Miami FL")
def onclick(location):
    geodata = geocode(location)
    latlon = geodata['results'][0]['geometry']['location']
    current = weather(latlon['lat'], latlon['lng'])
    print(current['current']['apparent_temperature'])
```
