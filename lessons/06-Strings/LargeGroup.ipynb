```python
def mask_credit_card( card: str) -> str:
    '''
    This function take a credit card like
    3145666329184490
    and masks it like this:
    XXXXXXXXXXXX4490
    '''
    last_four = card[-4:]
    remaining = card[:-4]
    digits = len(remaining)
    masked_card = "X" * digits + last_four
    return masked_card
```

```python
expected = 'XXXXXXXX9990'
actual = mask_credit_card("123456789990")
print("Expected:", expected, "Actual:", actual)
assert expected == actual
assert mask_credit_card("00000") == "X0000"
```
