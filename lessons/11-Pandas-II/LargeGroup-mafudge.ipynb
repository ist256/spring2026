```python
def games_played(wins: int, losses: int) -> int:
    '''
    this silly function takes two numbers wins and losses 
    and adds em up and returns them
    '''
    total = wins + losses
    return total

print(games_played(7,3) == 10)
```

```python
mlb["GP"] = mlb.apply( lambda row: games_played(row["W"], row["L"]), axis=1)
mlb
```
