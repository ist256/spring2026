```python
# Examples of highlighting points in plots
```

```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mafudge/datasets/master/exam-scores/exam-scores.csv")
df.sample(10)
```

```python
# axvline (Axis Vertial Line)
myscore = df.sample(1)['Student_Score'].values[0]
mean = df['Student_Score'].mean()
sns.histplot(data=df, x="Student_Score", kde=True)
plt.axvline(x = myscore,  color='red', linestyle='--', label=f'Your Score: {myscore}')
plt.axvline(x = mean,  color='lime', linestyle='solid', label=f'Mean: {mean:.1f}')
plt.legend()
plt.show()
```

```python
# Highlight points
student1 = df.sample(1)
student2 = df.sample(1)
sns.lmplot(data=df, x="Completion_Time", y="Student_Score", hue="Class_Section",legend=False,fit_reg=False)
plt.scatter(x=student1["Completion_Time"], y=student1['Student_Score'], 
            color='red', s=100, edgecolor='black', label=f'Student 1')
plt.scatter(x=student2["Completion_Time"], y=student2['Student_Score'], 
            color='lime', s=100, edgecolor='black', label=f'Student 2')
plt.legend()
plt.show()
```
