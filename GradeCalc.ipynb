```python
import pandas as pd
from IPython.display import display, HTML
from ipywidgets import interactive, IntSlider
grade_table = pd.DataFrame({"student_achievement": ["Mastery","Mastery","Satisfactory","Satisfactory","Satisfactory","Low Passing","Low Passing","Low Passing","Unsatisfactory","Unsatisfactory"],"total_points_earned": ["243-256","230-242","217-229","204-216","192-203","179-191","166-178","153-165","128-152","0-127"],"registrar_grade": ["A","A-","B+","B","B-","C+","C","C-","D","F"],"grade_points": [4.0,3.666,3.333,3.0,2.666,2.333,2.0,1.666,1.0,0.0]})
grade_table[["minpts","maxpts"]] = grade_table["total_points_earned"].str.split("-", expand=True).astype(int)
def lookup(pts):
    row = grade_table[(grade_table["minpts"] <= pts) & (pts <= grade_table["maxpts"])].iloc[0]
    return row["registrar_grade"], row["student_achievement"]
def onclick(earned, total):
    if earned > total:
        display(HTML("<h4>💣ERROR: 'Pts Earned' must be less than 'Pts So Far'</h4>"))
    else:
        proj = 256 * earned // total
        maxp = 256 - total + earned
        pg, pa = lookup(proj)
        mg, ma = lookup(maxp)
        display(pd.DataFrame({"Projected": [proj, pg, pa], "Maximum": [maxp, mg, ma]}, index=["Points","Grade","Achievement"]))
display(HTML("<H1>🧮 IST256 Grade Calculator</H1>"))
display(HTML('''<ul><li>The course maximum is <b>256</b> points.<li>Use the sliders to select your values or click on the number to type a value.<li>See syllabus for details: <a href="https://ist256.com/syllabus/#grading-scale-for-final-grade">https://ist256.com/syllabus/#grading-scale-for-final-grade</a>.<li><i>Projected</i> - Estimated final grade based on points so far.<li><i>Maximum</i> - Highest possible grade based on points remaining.</ul>'''))
display(interactive(onclick, earned=IntSlider(min=0,max=256,step=1,description="Pts Earned"), total=IntSlider(min=1,max=256,step=1,description="Pts So Far")))
```
